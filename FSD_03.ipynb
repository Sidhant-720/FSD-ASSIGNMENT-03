Q1:-Create a layout with three boxes side by side using flexbox. the first box should have twice the width of the other two. Provided the HTML and CSS code snippets.
* Ans:- Here's how we can create that layout with flexbox.

````python
HTML:

"html

<div class="container">

<div class="box box-1">Box 1</div>

<div class="box box-2">Box 2</div>

<div class="box box-3">Box 3</div>

</div>

CSS:

```CSS

.container {

}

display: flex; /* Enables Flexbox */

.box {

/* Basic styling for the boxes */

padding: 20px;

text-align: center;

border: 1px solid black;

}

.box-1 {

flex: 2; /* Box 1 takes up twice the space */}

.box-2,

.box-3 {

flex: 1; /* Boxes 2 and 3 share the remaining

space equally */
}
```
````

Explanation:

1. HTML Structure: We start with a container 'div'

that holds three 'div' elements, each representing a box.

2. CSS-Container:

*

`display: flex; This is the key. It turns the container into a flex container, enabling Flexbox.

3. CSS Boxes:

* `flex: 2; This is applied to`.box-1`. The `flex` property is a shorthand for 'flex-grow', `flex-shrink, and `flex-basis'. Here, `flex: 2` means this box will grow twice as much as the other boxes to fill available space.

* `flex: 1; This is applied to`.box-2` and `.box-3`. This means they'll share the remaining space equally.

This setup ensures that Box 1 occupies twice the width of Box 2 and Box 3, which will be side by side.

Q2:-Using CSS Grid , create a 2*2 grid layout where the first row spans two columns, and the second row has two equal columns. Provide the HTML and CSS code snippets.
* Ans:-

````python
HTML:

"html

<div class="container">

<div class="item item-1">Item 1</div>

<div class="item item-2">Item 2</div>

<div class="item item-3">Item 3</div>

<div class="item item-4">Item 4</div>

</div>

CSS:

```CSS

.container {

display: grid; /* Enables Grid */

grid-template-columns: repeat(2, 1fr); /* Two columns of equal width */grid-template-rows: auto auto; /* Two rows, height auto adjusts to content */}

.item {

padding: 20px;

border: 1px solid black;

text-align: center;

}

.item-1 {

grid-column: 1 / span 2; /* Spans across two columns */

}
```
````

Explanation:

1. HTML Structure: We have a container 'div with four 'div' elements, each representing a grid item.

2. CSS-Container:

* 'display: grid; This turns the container into a grid container.

* grid-template-columns: repeat(2, 1fr); This creates two columns. `repeat(2, 1fr) means "repeat the `1fr unit twice." The 'fr' unit represents a fraction of the available space, so each column gets an equal width.

* `grid-template-rows: auto auto; This creates two rows. `auto' means the row height will adjust to fit the content.

3. CSS-Item 1:

* `grid-column: 1/ span 2; This is the key to making the first item span two columns.

`grid-column: 1` starts the item at the first column line, and/ span 2' tells it to span across two columns.

Q6:-Create a simple animation that moves a square box from left to right infinitely. Provided the HTML and CSS code snippets.
* Ans:-

````python
HTML:

"html

<div class="container">

<div class="box"></div>

</div>
```

CSS:

```CSS

.container {

width: 100%;

overflow: hidden; /* Prevents the box from

overflowing */

}

.box {

width: 50px;

height: 50px;

background-color: blue;

position: relative; /* Required for animation */

animation: move Right 3s linear infinite; /*

Applies the animation */

}

@keyframes move Right {

0% {

left: 0; /* Start position */

100% {

 left: calc(100% - 50px); /* End position (right }

edge) */

}
}
```
````

Q7:-Explain the difference between `position:absolute;` and `position:fixed;`in CSS.
* Ans:-  `position: absolute;`

* Positioning Context: An element with `position: absolute; is positioned relative to its *closest position ancestor*.

* A "positioned ancestor" is an ancestor element (parent, grandparent, etc.) that has a `position` value other than 'static' (which is the default). This means the ancestor has 'position: relative;', 'position: absolute;', 'position: fixed;', or 'position: sticky;`.

* If there's no positioned ancestor, the absolutely positioned element is positioned relative to the initial containing block, which is usually the <html> or <body>` element.

* How it Works: The element is removed from the normal document flow. Space is *not* created for the element in the layout. Other content will flow as if the absolutely positioned element doesn't exist. You then use 'top', 'right', 'bottom', and 'left' properties to specify the element's position relative to its positioned ancestor.

* Use Cases:

* Overlapping elements.

* Creating elements that are positioned relative to a specific container, regardless of the rest of the page content.

* Complex layouts where you need precise control over element placement.
-- `position: fixed;`

* Positioning Context: An element with 'position: fixed; is positioned relative to the *browser window* (viewport). It stays in the same position even when the user scrolls the page.

* How it Works: Like 'absolute', the element is

removed from the normal document flow. Space is *not* created for the element in the layout. The element's position is determined by the 'top', 'right', 'bottom', and 'left' properties, relative to the viewport.

* Use Cases:

* Creating a sticky header or navigation bar that stays at the top of the screen.

* Creating a floating button or element that always remains visible.

* Displaying chat windows or other Ul elements that should always be accessible.
